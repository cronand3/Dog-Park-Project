```python
# deal with the yucky date/time string
# this function will create an array with the correct times in military time in the format [hour, minute, second]
def correct_time(array):
    correct_times = []
    for i in range(0,len(array)):
        date_time = array[i][2] 
        time = date_time[-10:-2]
        am_pm = date_time[-2]
        minute = int(time[3:5])
        second = int(time[6:8])
        hour = int(time[0:2]) 
        if am_pm == 'A' and hour != 12:
            correct_times.append([hour, minute,second])
        elif am_pm == 'P'and hour != 12:
            hour += 12
            correct_times.append([hour, minute, second])
        elif am_pm == 'A' and hour == 12:
            hour = 0
            correct_times.append([hour, minute, second])
        elif am_pm == 'P' and hour == 12:
            hour = 12
            correct_times.append([hour,minute,second])
    return(correct_times)    
```

```python
# dealing with the yucky date/time string continued
# this function will create an array with the correct date in the format [month, day, year]
def correct_date(array):
    correct_dates = []
    for i in range(0, len(array)):
        date_time = array[i][2]
        date = date_time[0:10]
        find_month = date.partition('/')
        month = int(find_month[0])
        find_day = find_month[2].partition('/')
        day = int(find_day[0])
        year = int(find_day[2])
        correct_dates.append([month,day,year])
    return(correct_dates)
```

```python
def find_indices_per_day (length_of_month, month_num, year_num, date_array):
    
    
    # finds the indices in the dataset of each day in the given input month 
    i = 1
    month_year_list = []
    while i <= (length_of_month-1):
        begin1, end1 = find_indices(month_num,i,year_num,date_array)
        beginend = [begin1, end1]
        month_year_list.append(beginend)
        i+=1
    return month_year_list

def find_unique_ids (month_year_list, length_of_month, month_num, year_num, date_array):
 # add all the unique ID's to a new list of strings for each month
    userid_monthyear = []
    total_month = []
    for t in range(length_of_month-1):
        for i in range(0,len(date_array)):
            userid_monthyear.append(date_array[i][1])
        temp = userid_monthyear[month_year_list[t-1][0]:month_year_list[t-1][1]]
        total_month.append(temp)
        t+=1
    return total_month

def find_daily_frequencies (total_month, length_of_month, month_num, year_num, date_array):
 # find daily frequencies
    dailyfreq_ = []
    dailyuserid = []

    for i in range(length_of_month):
        dailyuserid = total_month[i-1]
        dailyuniqueusers = np.unique(dailyuserid)
        dailyfreq_.append(len(dailyuniqueusers))
    # create a list of all the possible dates 1 - length of the mongth     
    date = []

    for i in range(length_of_month):
        i+=1
        date.append(i)

    
    # find day that had the most frequencies and what index that date was
    max_val_month_year = max(dailyfreq_)
    max_index_month_year = dailyfreq_.index(max_val_month_year)
    print(max_index_month_year)
    
    # use datetime and weekday() to get an integer to determine which day of the week it is
    day_of_week = dt.date(year_num, month_num, (max_index_month_year+1)).weekday()

    if day_of_week == 0:
        print('Monday')
    elif day_of_week == 1:
        print("Tuesday")
    elif day_of_week == 2:
        print("Wednesday")
    elif day_of_week == 3:
        print("Thursday")
    elif day_of_week == 4:
        print("Friday")
    elif day_of_week == 5:
        print("Saturday")
    else:
        print("Sunday")
        
    # plot the unique scans by day for the given month
    
    plt.plot(date, dailyfreq_, color = 'blue')
    plt.grid()
    plt.title("Unique Scans by Day")
    plt.xlabel("Date")
    plt.xticks(np.arange(min(date), max(date)+1, 1.0))
    plt.ylabel("Total Scans")
    plt.ylim(0,70)
    plt.xlim(1,length_of_month)
    #plt.figure(figsize=(8,12))
    plt.rcParams['figure.figsize'] = [10, 5]
    plt.show()
    print(dailyfreq_)
        
        
    return dailyfreq_, dailyuniqueusers, dailyuserid

```
